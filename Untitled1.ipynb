```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Data_csv.csv')

window_size = 10  # Adjust the window size according to your data
plt.plot(data['x'],data['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('unfiltered')
plt.show()
filtered_data = data['y'].rolling(window_size, center=True).mean()
plt.plot(data['x'],filtered_data)
plt.xlabel('x')
plt.ylabel('y')
plt.title('filtered')
plt.show()
```
